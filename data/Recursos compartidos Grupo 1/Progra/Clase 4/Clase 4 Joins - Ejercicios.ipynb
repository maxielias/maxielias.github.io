### Ejercicio 1

Utilizando los `DataFrames` declarados, obtener los nombres de los empleados que trabajan en el area de venta (`Sales`). 

**Tips de funciones a utilizar:**
  - merge
 ``` python
 DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)
 ```

````python
### Completa con tu código aquí
## Acá hago el merge con company y area como keys (on)
df_merge = pd.merge(employee_df, company_df, how='left', left_on=['company', 'area'], right_on=['company_id', 'area_id'])

## Acá creo la máscara que voy a usar para filtrar
maskSales = df_merge['area_y'] == 'Sales'

### Acá aplico la máscara y del DF resultante, sólo traigo el campo de 'name' 
df_merge[maskSales][['name']]#.count()/df_merge[['name']].count()
# display(df_merge.count())
# help(pd.merge)
###
````

### Ejercicio 2

Obtener las habilidades y el nombre del empleado con `id=36053225378`


**Tips de funciones a utilizar:**
  - merge
  - group_by
  - agg

````python
### Completa con tu código aquí
display(skills_df)
# display(skills_df.groupby('group').groups)
### En esta línea agrupo y aplico una lambda sobre la columna 'skills' para concatenar todos los valores en un mismo campo
## De esta manera, evito que el resultado del join tenga filas duplicadas 
# groupedSkills = skills_df.groupby('group')['skills'].apply(lambda x: ', '.join(x))
groupedSkills = skills_df.groupby('group').agg({'skills': list})
#display(skills_df.groupby('group')['skills'])
#display(skills_df.groupby('group')[['skills']])
display(groupedSkills)
df_emp_skills = pd.merge(df_merge, groupedSkills, left_on='area_y', right_on='group')
maskEmployeeId = df_emp_skills['employee_id'] == 36053225378
display(df_emp_skills[maskEmployeeId][['name', 'skills']])
###
````
