### Ejercicio 6

* Generar una columna que contenga la hora de las llamadas

* Discretizar la hora de modo tal que se mapeen los rangos horarios a las siguientes categorías:

    - 0-5am - night
    - 5-12 - morning
    - 12-20 - afternoon
    - 20 - 0 evening
    
    
* Hacer una tabla pivot que muestre la duración media de las llamadas en función de la franja horaria y del mes

* Hacer una tabla pivot que muestre el número de llamadas en función de la franja horaria y del mes


```python
### Completa con tu código aquí
def mapRange(hour):
    if hour > '00:00' and hour <= '05:00':
        return "night"
    elif hour > '05:00' and hour <= '12:00':
        return "morning"
    elif hour > '12:00' and hour <= '20:00':
        return "afternoon"
    else:
        return "evening"

df['hora_llamada'] = df['date'].str.split(" ", expand=True)[1]
df['rango_horario'] = df['hora_llamada'].map(lambda x: mapRange(x))
display(df.pivot_table(index={'rango_horario', 'month'}, values='duration', aggfunc='mean'))
display(df.pivot_table(index={'rango_horario', 'month'}, aggfunc='count', values='item'))
```
