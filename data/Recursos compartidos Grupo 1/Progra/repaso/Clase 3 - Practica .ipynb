You can use the `get_group` method in Pandas to get a specific group from a `groupby` object that was created by grouping by more than one column.

Here's an example:

```
import pandas as pd

df = pd.DataFrame({
    'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
    'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
    'C': [1, 2, 3, 4, 5, 6, 7, 8],
    'D': [9, 10, 11, 12, 13, 14, 15, 16]
})

# Group by columns "A" and "B"
grouped = df.groupby(['A', 'B'])

# Get group with values "foo" in column "A" and "one" in column "B"
grouped.get_group(('foo', 'one'))
```

In this example, we first create a DataFrame with two columns "A" and "B". We then group the data by columns "A" and "B" using the `groupby` method. Finally, we use the `get_group` method to get the group with values "foo" in column "A" and "one" in column "B".
